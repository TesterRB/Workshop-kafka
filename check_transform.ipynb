```python
rename_maps = {
    2015: {
        'Happiness Rank': 'Happiness_Rank',
        'Happiness Score': 'Happiness_Score',
        'Economy (GDP per Capita)': 'GDP_per_Capita',
        'Family': 'Social_Support',
        'Health (Life Expectancy)': 'Life_Expectancy',
        'Trust (Government Corruption)': 'Government_Corruption'
    },
    2016: {
        'Happiness Rank': 'Happiness_Rank',
        'Happiness Score': 'Happiness_Score',
        'Economy (GDP per Capita)': 'GDP_per_Capita',
        'Family': 'Social_Support',
        'Health (Life Expectancy)': 'Life_Expectancy',
        'Trust (Government Corruption)': 'Government_Corruption'
    },
    2017: {
        'Happiness.Rank': 'Happiness_Rank',
        'Happiness.Score': 'Happiness_Score',
        'Economy..GDP.per.Capita.': 'GDP_per_Capita',
        'Family': 'Social_Support',
        'Health..Life.Expectancy.': 'Life_Expectancy',
        'Trust..Government.Corruption.': 'Government_Corruption'
    },
    2018: {
        'Overall rank': 'Happiness_Rank',
        'Country or region': 'Country',
        'Score': 'Happiness_Score',
        'GDP per capita': 'GDP_per_Capita',
        'Social support': 'Social_Support',
        'Healthy life expectancy': 'Life_Expectancy',
        'Freedom to make life choices': 'Freedom',
        'Perceptions of corruption': 'Government_Corruption'
    },
    2019: {
        'Overall rank': 'Happiness_Rank',
        'Country or region': 'Country',
        'Score': 'Happiness_Score',
        'GDP per capita': 'GDP_per_Capita',
        'Social support': 'Social_Support',
        'Healthy life expectancy': 'Life_Expectancy',
        'Freedom to make life choices': 'Freedom',
        'Perceptions of corruption': 'Government_Corruption'
    }
}
```

```python
final_cols = ['Happiness_Rank', 'Country', 'Happiness_Score', 'GDP_per_Capita',
              'Social_Support', 'Life_Expectancy', 'Freedom', 
              'Government_Corruption', 'Generosity']
```

```python
drop_cols = {
    2015: ['Standard Error', 'Dystopia Residual', 'Region'],
    2016: ['Region', 'Lower Confidence Interval', 'Upper Confidence Interval', 'Dystopia Residual'],
    2017: ['Whisker.high', 'Whisker.low', 'Dystopia.Residual'],
    2018: [],
    2019: []
}
```

```python
for year, df in dfs.items():
    df.drop(columns=[c for c in drop_cols[year] if c in df.columns], inplace=True, errors='ignore')
    df.rename(columns=rename_maps[year], inplace=True)
    df = df[[col for col in final_cols if col in df.columns]]
    
    dfs[year] = df
```

```python
df_2015, df_2016, df_2017, df_2018, df_2019 = dfs.values()
```

```python
df_all = df_all.sort_values(by=['Country', 'Year']).reset_index(drop=True)

# Aplicar interpolacin lineal solo dentro de cada pas
df_all['Government_Corruption'] = (
    df_all.groupby('Country', group_keys=False)['Government_Corruption']
          .apply(lambda x: x.interpolate(method='linear', limit_direction='both'))
)

missing = df_all['Government_Corruption'].isna().sum()
print(f"Valores faltantes restantes en Government_Corruption: {missing}")
```

```python
def get_country_name(name):
    try:
        # Busca el pas por nombre o por nombre comn
        country = pycountry.countries.lookup(name)
        return country.name
    except LookupError:
        return None  # Pas no reconocido
```

```python
df_all['Country_Standardized'] = df_all['Country'].apply(get_country_name)
```

```python
unmatched = df_all[df_all['Country_Standardized'].isna()]['Country'].unique()
print(f"Pases no reconocidos: {len(unmatched)}")
print(unmatched)
```

```python
manual_fix = {
    "Congo (Brazzaville)": "Republic of the Congo",
    "Congo (Kinshasa)": "Democratic Republic of the Congo",
    "Hong Kong S.A.R., China": "Hong Kong",
    "Ivory Coast": "Cte d'Ivoire",
    "Kosovo": "Republic of Kosovo",
    "Macedonia": "North Macedonia",
    "North Cyprus": "Northern Cyprus",
    "Northern Cyprus": "Northern Cyprus",
    "Palestinian Territories": "Palestine, State of",
    "Russia": "Russian Federation",
    "Somaliland Region": "Somaliland",
    "Somaliland region": "Somaliland",     
    "Swaziland": "Eswatini",
    "Taiwan Province of China": "Taiwan",
    "Trinidad & Tobago": "Trinidad and Tobago",
    "Turkey": "Trkiye"
}

df_all['Country_Standardized'] = (
    df_all['Country_Standardized']
    .fillna(df_all['Country'].map(manual_fix))
)
```

```python
remaining = df_all[df_all['Country_Standardized'].isna()]['Country'].unique()
print(f"Pases an sin estandarizar: {remaining}")
```

```python
df_all['Country'] = df_all['Country_Standardized']
df_all.drop(columns='Country_Standardized', inplace=True)
df_all.head()
```
